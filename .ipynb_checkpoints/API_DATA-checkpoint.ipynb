```python
import requests
import json
import time
import urllib.request
import numpy as np
key = '8NY5A51OROX11O2Y'
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol=IBM&interval=1min&apikey=8NY5A51OROX11O2Y'
producer = KafkaProducer(bootstrap_servers="localhost:9092",value_serializer=lambda v: json.dumps(v).encode('utf-8'))
topic_name = "cac40"


response = urllib.request.urlopen(url)
stations = json.loads(response.read().decode())
stations = stations.get('Time Series (1min)')
stations = pd.DataFrame(stations)
stations = stations.drop(['5. volume','2. high','3. low','1. open'],axis=0)
stations= stations.T
stations['date'] = stations.index
stations= stations.reset_index()
stations = stations.drop('index',axis=1)
stations['date'] = pd.to_datetime(stations['date'])
stations['date'] = stations[['date']].apply(lambda x: x[0].timestamp(), axis=1)

for indice in stations.index :
    message = json.dumps(dict(stations.loc[indice]),indent=2)
    print(message)
    producer.send(topic_name, message)

    #print("{} Produced {} station records".format(time.time(), len(stations)))
    #time.sleep(1)
   
```
